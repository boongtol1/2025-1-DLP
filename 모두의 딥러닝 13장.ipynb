## 3. 학습셋과 테스트셋

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split

import pandas as pd
```

```python
# 깃허브에 준비된 데이터를 가져옵니다. 앞에서 이미 데이터를 가져왔으므로 추석 처리합니다. 3번 예제만 별도 실행 시 주석을 해제하여 실습하세요.
!git clone https://github.com/taehojo/data.git
```

```python
# 광물 데이터를 불러옵니다.
df = pd.read_csv('./data/sonar3.csv', header=None)
```

```python
# 음파 관련 속성을 X로, 광물의 종류를 y로 저장합니다.
X = df.iloc[:,0:60]
y = df.iloc[:,60]
```

```python
# 학습셋과 테스트셋을 구분합니다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)

# 설정 방식:
# 비율: test_size=0.3 → 데이터의 30%가 테스트 데이터로 설정.
# 샘플 수: test_size=10 → 10개의 샘플을 테스트 데이터로 설정.
```

```python
# 모델을 설정합니다.
model = Sequential()
model.add(Dense(24,  input_dim=60, activation='relu'))
model.add(Dense(10, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# 모델을 컴파일합니다.
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델을 실행합니다.
history=model.fit(X_train, y_train, epochs=200, batch_size=10)
```

```python
# 모델을 테스트셋에 적용해 정확도를 구합니다.
score=model.evaluate(X_test, y_test)
print('Test accuracy:', score[1])

# score[0]: 테스트 데이터의 손실 값.
# score[1]: 테스트 데이터의 정확도.
```

## 5. k겹 교차 검증

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

import pandas as pd

# 깃허브에 준비된 데이터를 가져옵니다. 앞에서 이미 데이터를 가져왔으므로 추석 처리합니다. 3번 예제만 별도 실행 시 주석을 해제하여 실습하세요.
!git clone https://github.com/taehojo/data.git

# 광물 데이터를 불러옵니다.
df = pd.read_csv('./data/sonar3.csv', header=None)

# 음파 관련 속성을 X로, 광물의 종류를 y로 저장합니다.
X = df.iloc[:,0:60]

y = df.iloc[:,60]
print(y)
```

```python
# 몇 겹으로 나눌 것인지를 정합니다.
k=5

# KFold 함수를 불러옵니다. 분할하기 전에 샘플이 치우치지 않도록 섞어 줍니다.
kfold = KFold(n_splits=k, shuffle=True)

# 정확도가 채워질 빈 리스트를 준비합니다.
acc_score = []

def model_fn():
    model = Sequential() # 딥러닝 모델의 구조를 시작합니다.
    model.add(Dense(24, input_dim=60, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    return model

# K겹 교차 검증을 이용해 k번의 학습을 실행합니다.
for train_index , test_index in kfold.split(X):  # for 문에 의해서 k번 반복합니다. spilt()에 의해 k개의 학습셋, 테스트셋으로 분리됩니다.
    X_train , X_test = X.iloc[train_index,:], X.iloc[test_index,:]
    y_train , y_test = y.iloc[train_index], y.iloc[test_index]

    model = model_fn()
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    history=model.fit(X_train, y_train, epochs=200, batch_size=10, verbose=0)

    accuracy = model.evaluate(X_test, y_test)[1]  # 정확도를 구합니다.
    acc_score.append(accuracy)  # 정확도 리스트에 저장합니다.

# k번 실시된 정확도의 평균을 구합니다.
avg_acc_score = sum(acc_score)/k

# 결과를 출력합니다.
print('정확도:', acc_score)
print('정확도 평균:', avg_acc_score)
# k=5: 데이터를 5개의 부분으로 나눕니다. 한 번에 한 부분은 테스트셋으로, 나머지 부분은 학습셋으로 사용합니다. 의 의미는
# 전체 데이터를 여러 부분(폴드)로 나눈 뒤, 각 폴드가 한 번씩 테스트셋으로 사용되고 이때 나머지 폴드는 학습셋으로 사용된다는 뜻입니다.   // 1개짜리가 테스트 데이터다!
# shuffle=True: 데이터를 무작위로 섞은 뒤 나눕니다. 데이터가 특정 순서에 의존하지 않도록 방지합니다.
# 가중치 초기화 안할시 문제점1: 가중치가 초기화되지 않고 이전 폴드에서 학습된 가중치가 다음 폴드에 전달된다면, 일부 경우에는 테스트셋에 포함된 데이터의 정보를 모델이 이미 알고 있는 상태에서 학습 및 평가가 이루어질 수 있습니다.
# 이런 경우 모델의 성능(예: 정확도)이 더 좋아 보일 가능성이 있습니다. 하지만 이는 실제 모델의 일반화 성능을 반영하지 못하는 문제점을 야기합니다.
# 가중치 초기화 안할시 문제점2: 하지만 이전 폴드의 학습 과정에서 테스트셋의 정보가 모델에 반영되었다면, 이후 폴드에서 학습에 유리한 조건이 만들어질 수 있습니다.
# 가중치 초기화 안할시 문제점들의 공통점?--> "일반화" 의 어려움!!--> 손으로 써보면서 하는게 제일 좋음!
# K-겹 교차 검증에서 각 폴드마다 새로운 모델을 생성하기 위해 모델 생성 코드를 함수로 정의합니다.
# 함수로 정의하면 반복 과정에서 매번 새로운 모델을 생성할 수 있습니다.
# 결국, **“5번 반복하라”**는 로직은 KFold(n_splits=5) 설정에 이미 포함되어 있다고 이해하시면 됩니다.
# 네, 맞습니다. 예컨대 전체 데이터가 100개이고 이를 5개의 그룹(S1, S2, S3, S4, S5)으로 나눴다면, 한 그룹(S1)이 테스트셋으로 선정되었을때:
# test_index는 해당 그룹에 해당하는 20개 데이터의 인덱스를 담게 되고,
# train_index는 나머지 80개 데이터(S2, S3, S4, S5)의 인덱스를 담게 됩니다.
# 오름차순으로 정렬되어 출력(0,1,2,3,...9)된다는 것: 실제로 폴드에 속한 인덱스들을 반환할 때, scikit-learn이 관례적으로 정렬해서 주는 것.
# 네, 맞습니다. kfold.split(X) 는 결국 **테스트에 사용할 인덱스(test_index)**와 **학습에 사용할 인덱스(train_index)**를 튜플 형태로 반환합니다.
# verbose**는 출력되는 로그(log)의 상세 정도를 조절하는 옵션입니다.
```
